```python
### setting the index to 'release_date'
```

```python
df_tracks_indexed = df_tracks.set_index('release_date')
df_tracks_indexed.head()
```

```python
### changing column name 'duration_ms' to 'duration' with its values as well
```

```python
df_tracks_indexed["duration"] = df_tracks_indexed["duration_ms"].apply(lambda x: round(x/1000))
df_tracks_indexed.drop("duration_ms", inplace = True, axis = 1)
df_tracks_indexed.duration.head()
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'],axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt='.1g', vmax = 1, vmin = -1, center = 0, cmap = 'inferno', linewidths = 1, linecolor = 'Black')
heatmap.set_title("Correlation Heatmap btw Variables")
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))  # taking out sample data : considering only 0.4 percent of the whole data
len(sample_df)
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x = "energy", y = "loudness", color = "c").set(title = "Loudness vs Energy Correlation")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, x = "acousticness", y = "popularity", color = "g").set(title = "Popularity vs Acousticness Correlation")
```

```python
df_tracks_indexed['dates'] = df_tracks_indexed.index.get_level_values('release_date')    # used the already created variable in which the release date had been set to index
df_tracks_indexed.dates = pd.to_datetime(df_tracks_indexed.dates)
years = df_tracks_indexed.dates.dt.year
df_tracks_indexed.head()
```

```python
sns.displot(years, discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = "Number of songs released per year")
```

```python
total_dr = df_tracks_indexed.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
figure = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation = 90)
```

```python
total_dr = df_tracks_indexed.duration
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year vs Duration")
plt.xticks(rotation = 80)
```
